```python
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
female_survival_per = train_data.loc[train_data.Sex=='female']['Survived']
rate_female = sum(female_survival_per)/len(female_survival_per) 
print("% of women survived is : ", rate_female)
```

```python
male_survival_rate = train_data.loc[train_data.Sex=='male']['Survived']
rate_male = sum(male_survival_rate)/len(male_survival_rate)
print("% of male survived is :",rate_male)
```

Building RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error

y = train_data['Survived'] 
features = ['Pclass','Sex','SibSp','Parch']
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100,max_depth=5,random_state=1)
model.fit(X,y)
predictions = model.predict(X_test)
output = pd.DataFrame({'PassengerId':test_data.PassengerId,'Survived':predictions})
output.to_csv('my_submission.csv',index=False)
print("Your submission was successfully saved!")
```
